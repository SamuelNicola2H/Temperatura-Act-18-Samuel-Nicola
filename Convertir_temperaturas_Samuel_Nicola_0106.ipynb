## 1. Fórmulas de Conversión

Para convertir entre estas dos escalas, utilizamos las siguientes relaciones matemáticas:

* **De Celsius a Fahrenheit:**
    $$°F = (°C \times \frac{9}{5}) + 32$$
    *(También se puede usar $1.8$ en lugar de $9/5$)*

* **De Fahrenheit a Celsius:**
    $$°C = \frac{°F - 32}{1.8}$$
    *(O bien, multiplicar por $5/9$ después de restar 32)*

---

## 2. Ejemplos Resueltos Paso a Paso

### Ejemplo A: Convertir $25°C$ a $°F$ (Temperatura ambiente)
1.  **Identificar la fórmula:** $°F = (°C \times 1.8) + 32$
2.  **Sustituir el valor:** $°F = (25 \times 1.8) + 32$
3.  **Multiplicar:** $25 \times 1.8 = 45$
4.  **Sumar 32:** $45 + 32 = 77$
* **Resultado:** $25°C$ equivale a **$77°F$**.

### Ejemplo B: Convertir $98.6°F$ a $°C$ (Temperatura corporal)
1.  **Identificar la fórmula:** $°C = (°F - 32) / 1.8$
2.  **Sustituir el valor:** $°C = (98.6 - 32) / 1.8$
3.  **Restar 32:** $98.6 - 32 = 66.6$
4.  **Dividir entre 1.8:** $66.6 / 1.8 = 37$
* **Resultado:** $98.6°F$ equivale a **$37°C$**.

## 3. Código en Python

Aquí tienes el script interactivo siguiendo tus requerimientos técnicos. He añadido un manejo básico de errores por si el usuario ingresa letras en lugar de números.

```python
def convertirCF():
    celsius = float(input("\nIngresa los grados Celsius a convertir: "))
    fahrenheit = (celsius * 1.8) + 32
    print(f"---> {celsius}°C es igual a {fahrenheit:.2f}°F")

def convertirFC():
    fahrenheit = float(input("\nIngresa los grados Fahrenheit a convertir: "))
    celsius = (fahrenheit - 32) / 1.8
    print(f"---> {fahrenheit}°F es igual a {celsius:.2f}°C")

def menu():
    while True:
        print("\n--- MENÚ DE CONVERSIONES ---")
        print("1. Convertir Celsius a Fahrenheit")
        print("2. Convertir Fahrenheit a Celsius")
        print("3. Salir")
        
        opcion = input("Selecciona una opción (1-3): ")

        if opcion == '1':
            try:
                convertirCF()
            except ValueError:
                print("Error: Por favor ingresa un número válido.")
        
        elif opcion == '2':
            try:
                convertirFC()
            except ValueError:
                print("Error: Por favor ingresa un número válido.")
        
        elif opcion == '3':
            print("Saliendo del programa... ¡Hasta luego!")
            break
        
        else:
            print("Opción no válida, intenta de nuevo.")

# Ejecutar el programa
if __name__ == "__main__":
    menu()
```

### Explicación del Código:
* **`while True`**: Crea un bucle infinito para que el menú se repita hasta que el usuario decida salir.
* **`float()`**: Permite que el usuario ingrese números decimales (como 36.5).
* **`f"{valor:.2f}"`**: Es un formato para que el resultado solo muestre 2 decimales y sea más fácil de leer.
* **`try/except`**: (Opcional pero recomendado) Evita que el programa se cierre si el usuario escribe texto en lugar de números.
* **`break`**: Rompe el ciclo `while` inmediatamente cuando se selecciona la opción 3.

````python
def convertirCF():
    """Convierte grados Celsius a Fahrenheit"""
    celsius = float(input("\nIngrese los grados Celsius a convertir: "))
    fahrenheit = (celsius * 1.8) + 32
    print(f">> {celsius}°C es igual a {fahrenheit:.2f}°F")

def convertirFC():
    """Convierte grados Fahrenheit a Celsius"""
    fahrenheit = float(input("\nIngrese los grados Fahrenheit a convertir: "))
    celsius = (fahrenheit - 32) / 1.8
    print(f">> {fahrenheit}°F es igual a {celsius:.2f}°C")

def menu():
    """Muestra el menú principal y gestiona la lógica"""
    while True:
        print("\n--- CONVERSOR DE TEMPERATURA ---")
        print("1. Convertir Celsius a Fahrenheit")
        print("2. Convertir Fahrenheit a Celsius")
        print("3. Salir")

        opcion = input("Seleccione una opción (1-3): ")

        if opcion == "1":
            convertirCF()
        elif opcion == "2":
            convertirFC()
        elif opcion == "3":
            print("Saliendo del programa... ¡Hasta luego!")
            print("Programa realizado por Samuel Nicola NL = 40")
            break
        else:
            print("Opción no válida, por favor intente de nuevo.")

# Punto de entrada del programa
if __name__ == "__main__":
    menu()
````
